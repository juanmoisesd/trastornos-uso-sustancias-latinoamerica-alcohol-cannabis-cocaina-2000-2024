## 1. Configuración

````python
!pip install pandas matplotlib seaborn numpy -q
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
plt.style.use("seaborn-v0_8-whitegrid")
print("✅ Ready")
````

## 2. Datos de muestra

````python
# Descargar desde Zenodo: https://github.com/juanmoisesd/trastornos-uso-sustancias-latinoamerica-alcohol-cannabis-cocaina-2000-2024
np.random.seed(42)
countries = ["Brazil","Mexico","Argentina","Colombia","Chile","Peru","Venezuela","Ecuador"]
years = list(range(2015, 2026))
rows = []
for c in countries:
    base = np.random.uniform(8,25)
    for i,y in enumerate(years):
        rows.append({"pais":c,"anio":y,"value":round(base+i*np.random.uniform(0.1,0.8)+np.random.normal(0,0.5),2),"n":int(np.random.uniform(500,5000))})
df = pd.DataFrame(rows)
print(f"Shape: {df.shape}")
df.head()
````

## 3. Visualización

````python
fig,axes=plt.subplots(1,2,figsize=(14,5))
for c in countries:
    s=df[df["pais"]==c]
    axes[0].plot(s["anio"],s["value"],marker="o",ms=3,label=c)
axes[0].set_title("Tendencia por País",fontsize=13,fontweight="bold")
axes[0].legend(fontsize=7,ncol=2)
pivot=df.pivot_table(values="value",index="pais",columns="anio",aggfunc="mean").round(1)
sns.heatmap(pivot,ax=axes[1],cmap="YlOrRd",annot=True,fmt=".1f",annot_kws={"size":7},linewidths=0.5)
axes[1].set_title("Mapa de Calor",fontsize=13,fontweight="bold")
plt.tight_layout()
plt.savefig("example_plot.png",dpi=150,bbox_inches="tight")
plt.show()
````
